```python
file4 = uproot.open("zz2ll_delphes.root")
file4.keys()
```

```python
# Display the structure of zz2ll_delphes.root
print("Top-level keys in zz2ll_delphes.root:")
keys = list(file4.keys())
print(keys)

# Safely get the first tree and show its structure (avoid IndexError if file is empty)
if not keys:
	print("No top-level keys found in zz2ll_delphes.root.")
else:
	first_tree_name = keys[0]
	tree4 = file4[first_tree_name]
	print(f"\nStructure of tree '{first_tree_name}':")
	try:
		tree4.show()
	except Exception as e:
		print("Could not display the structure of the object:", e)
```
